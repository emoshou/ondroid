```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Excel file
data_path = 'matData.csv'  # Update with your file path
df = pd.read_csv(data_path)
```

```python
# Extracting relevant columns for the chart
material_names = df['material']
yield_strengths = df['yield']
costs = df['cost']
young = df['youngModulus']
density = df['density']

# Calculate Specific Strengths
specificStrengths = yield_strengths / density
```

```python
# Create the scatter plot using Matplotlib
plt.figure(figsize=(12, 8))

# Plot each point with a unique label
for material, x, y in zip(material_names, costs, yield_strengths):
    plt.scatter(x, y, label=material)

# Create the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xlabel('Cost')
plt.ylabel('Yield Strength (ksi)')
plt.title('Yield Strength vs. Cost of Metals from SendCutSend')
plt.grid(True)
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Create the scatter plot using Matplotlib
plt.figure(figsize=(12, 8))

# Plot each point with a unique label
for material, x, y in zip(material_names, costs, young):
    plt.scatter(x, y, label=material)

# Create the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xlabel('Cost')
plt.ylabel("Young's Modulus (E^6 PSI)")
plt.title("Young's Modulus (E^6 PSI) vs. Cost of Metals from SendCutSend")
plt.grid(True)
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Create the scatter plot
plt.figure(figsize=(12, 8))

# Plot each point with a unique label
for material, x, y in zip(material_names, costs, specificStrengths):
    plt.scatter(x, y)
    plt.text(x, y, material, fontsize=9, ha='left')

plt.xlabel('Cost')
plt.ylabel('Specific Strength (kip - in / lb)')
plt.title("Specific Strength (kip - in / lb) vs. Cost ($/lb) of Metals from SendCutSend")
plt.grid(True)
plt.tight_layout()

# Display the chart
plt.show()
```
