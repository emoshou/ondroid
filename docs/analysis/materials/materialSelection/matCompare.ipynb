```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Excel file
data_path = 'matData.csv'  # Update with your file path
df = pd.read_csv(data_path)
```

```python
# Extracting relevant columns for the chart
material_names = df['material']
yield_strengths = df['yield']
costs = df['cost']
```

```python
# Create the scatter plot using Matplotlib
plt.figure(figsize=(12, 8))

# Plot each point with a unique label
for material, x, y in zip(material_names, costs, yield_strengths):
    plt.scatter(x, y, label=material)

# Create the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xlabel('Cost')
plt.ylabel('Yield Strength (ksi)')
plt.title('Yield Strength vs. Cost of Various Materials from SendCutSend')
plt.grid(True)
plt.tight_layout()

# Display the chart
plt.show()
```
